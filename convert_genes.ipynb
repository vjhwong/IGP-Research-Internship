```python
import os
import pandas as pd
import numpy as np

from Ensembl_converter import EnsemblConverter

# Create an instance of EnsemblConverter
converter = EnsemblConverter()
```

```python
expression = pd.read_csv('GSE39582/GSE39582_Expression.txt', sep='\t')
expression = expression.rename(columns={'Unnamed: 0': 'SampleID'})
expression.head()
```

```python
if os.listdir('convert_dataframes'):
    start_index = int("".join([i for i in os.listdir('convert_dataframes')[0] if i .isdigit()]))
else:
    start_index = 0

# Divide the columns into batches of 200
column_batches = [expression.columns[i:i+200] for i in range(start_index, len(expression.columns), 200)]

# Process each batch and save the intermediate results
for i, batch in zip(np.arange(start_index, len(expression.columns), 200),column_batches):
    print(i)
    if os.listdir('convert_dataframes'):
        checkpoint_dataframe = pd.read_csv('convert_dataframes/Expression_batch_' + str(i) + '.csv')
    else:
        checkpoint_dataframe = pd.DataFrame()
    selected_expression = expression[batch]
    selected_expression = selected_expression.set_axis(converter.convert_ids(selected_expression.columns.to_list())["Symbol"], axis=1)
    checkpoint_dataframe = pd.concat([checkpoint_dataframe, selected_expression], axis=1)
    
    checkpoint_dataframe.to_csv('convert_dataframes/Expression_batch_' + str((i+200)) + '.csv', index=False)
    os.remove('convert_dataframes/Expression_batch_' + str(i) + '.csv')
```
